```python
import numpy as np

X = np.array([1,2,3,4]).reshape(-1,1)
Y = np.array([2,4,5,9]).reshape(-1,1)

X_design = np.hstack([np.ones((X.shape[0],1)), X])

beta = np.linalg.inv(X_design.T @ X_design) @ X_design.T @ Y
print("Intercept:", beta[0][0])
print("Slope:", beta[1][0])
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('canada_per_capita_income.csv')

X = df[['year']].values
Y = df.iloc[:,1].values.reshape(-1,1)

X_design = np.hstack([np.ones((X.shape[0],1)), X])

beta = np.linalg.inv(X_design.T @ X_design) @ X_design.T @ Y

pred = np.array([1,2020]) @ beta
print("Predicted income:", pred[0])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv('hiring.csv')

word_to_num = {
    'zero':0,
    'one':1,
    'two':2,
    'three':3,
    'four':4,
    'five':5,
    'six':6,
    'seven':7,
    'eight':8,
    'nine':9,
    'ten':10,
    'eleven':11,
    'twelve':12
}

df['experience'] = df['experience'].map(word_to_num)

df['experience'] = df['experience'].fillna(0)

df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(
    df['test_score(out of 10)'].median()
)

X = df[['experience',
        'test_score(out of 10)',
        'interview_score(out of 10)']]

y = df['salary($)']

model = LinearRegression()
model.fit(X, y)

pred1 = model.predict([[2,9,6]])
pred2 = model.predict([[12,10,10]])

print("Salary Prediction 1:", pred1[0])
print("Salary Prediction 2:", pred2[0])
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('hiring.csv')

mapping = {'zero':0,'one':1,'two':2,'three':3,'four':4,
           'five':5,'six':6,'seven':7,'eight':8,'nine':9,'ten':10}

df['experience'] = df['experience'].map(mapping).fillna(0)
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(
                                df['test_score(out of 10)'].median())

X = df[['experience','test_score(out of 10)','interview_score(out of 10)']].values
Y = df['salary($)'].values.reshape(-1,1)

X_design = np.hstack([np.ones((X.shape[0],1)), X])

beta = np.linalg.inv(X_design.T @ X_design) @ X_design.T @ Y

sample = np.array([1,12,10,10])
prediction = sample @ beta

print(prediction)
```
